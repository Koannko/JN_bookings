```python
def rename_cols(df):
    new_cols = {}
    for col in df.columns:
        new = col.replace(" ", "_").lower()
        new_cols[col] = new
    return new_cols

df = df.rename(columns=rename_cols(df))
df
```

```python
df[['adults', 'children', 'babies']].mean()
```

```python
info = df.groupby(['hotel', 'arrival_date_year', 'arrival_date_month'], as_index=False). \
    agg({'adults': 'mean', 'children': 'mean', 'babies': 'mean'}). \
    round(3)

total_col = info[['adults', 'children', 'babies']].sum(numeric_only=True, axis=1)

info['total_mean'] = total_col

info.sort_values(by=[ 'hotel', 'arrival_date_year', 'total_mean'], ascending=False)
```

```python
info_kids = df.groupby(['hotel'], as_index=False). \
    agg({'children': 'mean', 'babies': 'mean'}). \
    round(2)

total_kids_col = info_kids[['children', 'babies']].sum(numeric_only=True, axis=1)

info_kids['total_kids'] = total_kids_col

info_kids.sort_values('total_kids', ascending=False)
```
